```python
import ast
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movie['genres']=movie['genres'].apply(convert)
```

```python
def castconvert(obj1):
  count=0;
  L=[]
  for i in ast.literal_eval(obj1):
    if(count<3):
       L.append(i['name'])
       count+=1
    else:
      break
  return L
```

```python
movie['cast']=movie['cast'].apply(castconvert)
```

```python
def fetch_director(obj3):
  L=[]
  for i in ast.literal_eval(obj3):
    if(i['job']=='Director'):
      L.append(i['name'])
      break
  return L
```

```python
movie['crew']=movie['crew'].apply(fetch_director)
```

```python
movie['genres']=movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie['keywords']=movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie['cast']=movie['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie['crew']=movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie['overview']=movie['overview'].apply(lambda x: x.split())
```

```python
movie['tags']=movie['overview']+movie['genres']+movie['keywords']+movie['cast']+movie['crew']
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
a=['loves','loved','loving','lovable']
```

```python
ps=PorterStemmer()
```

```python
ps.stem('hatching')
```

```python
def stemming(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
stemming("loves loving lovable love")
```
